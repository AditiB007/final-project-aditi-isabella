# Top 100: Global on Apple Music - Web Scraper

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
response = requests.get("https://music.apple.com/us/playlist/top-100-global/pl.d25f5d1181894928af76c85c967f8f31")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
songList = soup.find_all("div", class_="songs-list-row songs-list-row--web-preview web-preview songs-list-row--two-lines songs-list-row--song")

ranks = []
songNames = []
artistNames = []
songTimes = []

for songs in songList:
    rank = songs.find("div", class_="songs-list__col songs-list__col--song typography-body").find("div", class_="songs-list-row__rank").text
    ranks.append(rank)
    song = songs.find("div", class_="songs-list__col songs-list__col--song typography-body").find("div", class_="songs-list-row__song-name").text
    songNames.append(song)
    artist = songs.find("div", class_="songs-list__col songs-list__col--artist typography-body").find("div", class_="songs-list__song-link-wrapper").find("a").text
    artistNames.append(artist)
    time = songs.find("div", class_="songs-list__col songs-list__col--time typography-body").find("div", class_="songs-list-row__length").text.strip()
    songTimes.append(time)
    # print(rank, song)
    # print(artist)
    # print(time)
    # print("------------------")
    # print()
    
list_of_tuples = list(zip(songNames, artistNames, songTimes))
df = pd.DataFrame(list_of_tuples, columns=["Song", "Artist", "Time"], index=ranks)
df
```

```python
df.to_csv('Apple_Music_Global_Top_100.csv')
```
