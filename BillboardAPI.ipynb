```python
import billboard
chart = billboard.ChartData('hot-100')
chart.title = 'The Hot 100'
print(chart)
```

```python
print(chart[50].title) #name of song
print(chart[50].artist) #artist of song
print(chart[50].rank) #current postion for the week
print(chart[50].peakPos) #peak rank
print(chart[50].lastPos) #postion from lest week
print(chart[50].weeks) # how many weeks in chart
print(chart[50].isNew) #newly entered to chart?
```

```python
rows = []
#for songs in chart:
for i in range(0,100):
        titl = (chart[i].title) #name of song
        art = (chart[i].artist) #artist of song
        rank = (chart[i].rank) #current postion for the week
        pRank = (chart[i].peakPos) #peak rank
        lRank =(chart[i].lastPos) #postion from lest week
        week = (chart[i].weeks) # how many weeks in chart
        new = (chart[i].isNew) #newly entered to chart?
        rows.append({
            "Title": titl,
            "Artist": art,
            "Current Position": rank ,
            "Last Week Position": lRank ,
            "Peak Position": pRank ,
            "Weeks On Chart": week ,
            "New to Chart?": new ,
    })
print(rows[:5])
```

```python
import pandas as pd
music_df = pd.DataFrame(rows)
#music_df.iloc[: , 1:]
music_df.head(20)
```

```python
music_df.to_csv('Billboard_Hot_100.csv')
```
